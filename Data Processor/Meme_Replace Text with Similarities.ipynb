```python
import pandas as pd
import numpy as np
import os
import augly.text as textaugs
import random
```

```python
initial_train=pd.read_json("C:\\Users\\yshen\\Documents\\Georgia Tech\\4\\Deep Learning\\pp_textadd\\hateful_memes\\train.jsonl",lines=True)
```

```python
initial_train.head()
```

```python
initial_train[(initial_train['label']==1)]
```

```python
num_alter=len(initial_train)
l=random.sample(range(0, num_alter), num_alter//2)
```

```python
aug=textaugs.ReplaceSimilarUnicodeChars(aug_word_p=0.8)
```

```python
for i in l:
    
    meta=[]
    t=aug(initial_train['text'][i], metadata=meta)
    
    df= pd.DataFrame([[initial_train['id'][i],initial_train['img'][i],initial_train['label'][i],t]],columns=['id','img','label','text'])
    
    initial_train=initial_train.append(df,ignore_index=True)
```
